```python
top_words = 10000 #Vocablury size
t = tf.keras.preprocessing.text.Tokenizer(num_words = top_words) # num_words -> Vocablury size
```

# Prepare Training and Test Data

Get the word index for each of the word in the review

```python
X_train[0:1] # view first data after tokenization
```

```python
X_train = t.texts_to_sequences(X_train.tolist())
```

```python
X_train[0:1][0]
```

```python
X_test = t.texts_to_sequences(X_test.tolist())  # coverting text to sequence of numbers
```

# Pad Sequences - Important

```python
#Define maximum number of words to consider in each review
max_review_length = 300
```

```python
#Pad training and test reviews
X_train = tf.keras.preprocessing.sequence.pad_sequences(X_train,maxlen = max_review_length,padding = 'pre')
X_test = tf.keras.preprocessing.sequence.pad_sequences(X_test,maxlen = max_review_length,padding = 'pre')
```

# Build the Graph

```python
#Initialize model
tf.keras.backend.clear_session()
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Embedding(top_words +1,50,input_length=max_review_length))  #Vocablury size = top_words,#Embedding size.every word lengh is of 50 numbers.#Number of words in each review
```

```python
model.output
```

Add LSTM Layer with 256 as RNN state size

```python
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dropout(0.5))
```

```python
model.add(tf.keras.layers.LSTM(256,dropout = 0.4)) #RNN State - size of cell state and hidden state is 256
```

```python
model.output
```

Use Dense layer for output layer

```python
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dropout(0.5))
```

```python
model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
```

```python
#Compile the model
model.compile(optimizer='adam',loss = 'binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

# Execute the graph

```python
from tensorflow.keras.callbacks import EarlyStopping
# This callback will stop the training when there is no improvement in the loss for three consecutive epochs.
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='loss',patience=5,mode="auto")  
```

```python
model.fit(X_train,y_train,epochs = 100,batch_size = 64,validation_data=(X_test,y_test),callbacks=[early_stopping])
```
